## Tugas 1 - Algoritma Valid

Perhatikan masalah berikut.

Misalkan terdapat $N$ buah pot bunga yang disusun secara melingkar dan masing-masing dilabeli dengan nomor $0,1,\dots, N-1$. Ronald bertugas untuk menyirami tanaman dalam pot tersebut, namun ia melakukannya dengan cara yang nyentrik. Pertama dia mulai dengan menyirami tanaman pada pot $0$. Kemudian, tanaman selanjutnya yang ia sirami adalah yang berjarak $M$ dari pot sebelumnya. Ia terus melakukan proses ini dalam 'sirkuit' pot bunga yang melingkar tersebut. Jika Ronald terus melakukan hal ini, ada berapa pot bunga di antaranya yang tidak pernah kena siraman?

Sebagai contoh, jika $N=8$ dan $M=2$, maka pot bunga yang disiram Ronald secara berurutan memiliki label $0, 2, 4, 6, 0, 2, 4, 6$ dst terus berulang. Ini menyisakan pot berlabel $1, 3, 5, 7$ yang tidak pernah kena siraman. Dengan demikian, jawaban yang diharapkan adalah $4$.

Berikut ini merupakan beberapa algoritma yang mencoba menyelesaikan masalah di atas dengan **tidak valid**.

**Batasan Input:**
* $N$ dan $M$ merupakan bilangan bulat positif.
* $M < N$.

**Tugas:**

Identifikasilah permasalahan/isu dari masing-masing algoritma di bawah ini.

````python
# ALGORITMA 1.1

[N, M] = list(map(int, input().split(" ")))

pot = [False]*N           #list berukuran N berisi False untuk menyimpan informasi mengenai pot mana saja yang sudah disiram
pos = 0                   #variabel yang menunjukkan posisi penyiraman saat ini

while (pos < N):
  pot[pos] = True         #siram pot pada posisi pos
  pos += M                #pindah ke posisi pos + M

counter = 0
for i in range(N):
  if (pot[i] == False):   #menghitung berapa banyak yang belum disiram
     counter += 1
print(counter)
````

**Algoritma 1.1 tidak valid**
* Algoritma tersebut tidak memiliki sifat sirkular sehingga output yang dihasilkan tidak selalu benar untuk semua N. Algoritma tersebut hanya menghitung untuk 1 kali putaran.
* Challenge: N = 8, M = 3 menghasilkan output = 5 -> harusnya output = 0.

````python
# ALGORITMA 1.2

[N, M] = list(map(int, input().split(" ")))

def PernahDisiram(i, N, M):
  # fungsi ini mencari tahu apakah pot ke i pernah disiram atau tidak

  pot = [False]*N #list berukuran N berisi False untuk menyimpan informasi mengenai pot mana saja yang sudah disiram
  pos = 0 #variabel yang menunjukkan posisi penyiraman saat ini
  while (pot[i] == False): #akan mencari kemungkinan pot ke i kena siram
    pot[pos] = True #siram pot pada posisi pos
    if pos==i: #pot ke i sudah tersiram
      return True

    pos = (pos + M) % N # pindah ke posisi pos + M
  return False

counter = 0
for i in range(N):
  if not PernahDisiram(i, N, M):
    counter+=1

print(counter)
````

**Algoritma 1.2 tidak valid**
* Dapat terjadi perulangan tak berhingga bila pot tidak tersiram semua.
* Challenge: N = 4 dan M = 2 menghasilkan infinite loop -> harusnya output = 2 
* Karena saat i = 1, maka pot[i] pasti akan selalu false. Hal ini karena perintah pot[pos] = True tidak akan terjadi pada pot[i] (disebabkan oleh pos tidak akan = 1, melainkan hanya bisa menghasilkan angka 0 dan 2)

## Tugas 2 - Algoritma Efisien

Diberikan bilangan bulat positif $N$ dan $K$ sebagai input. Sebuah matriks berukuran $N\times N$ berisi bilangan-bilangan terurut yang dituliskan dari kiri ke kenan, dan kemudian dari atas ke bawah. Algoritma diminta untuk mengoutputkan posisi baris dan kolom dari bilangan $K$ pada matriks tersebut.

Contoh, misalkan $N=3$ dan $K=7$. Maka matriks akan berisi elemen demikian:
```
1 2 3
4 5 6
7 8 9
```
Dapat dilihat bahwa posisi $K=7$ berada pada baris ke-3 dan kolom ke-1. Maka, output yang di harapkan adalah ```R = 3, C = 1```.

**Batasan Input:**
* $N$ dan $K$ merupakan bilangan bulat positif.
* $M < N$.

Di bawah ini terdapat beberapa algoritma valid untuk menjawab masalah ini yang Anda perlu lakukan analisis.

````python
# ALGORITMA 2.1

N = int(input("N = "))
K = int(input("K = "))

def search_matrix_1(N, K):
  counter = 1
  matrix = []

  # building the matrix
  for i in range(N):
    # building the i-th row of the matrix
    row = []
    for j in range(N):
      # building each element of the i-th row
      row.append(counter)
      counter += 1
    # append the i-th row of the matrix
    matrix.append(row)

  # looking for the position of K in the matrix
  for i in range(N):
    for j in range(N):
      if matrix[i][j] == K:
        return "R = " + str(i+1) + " C = " + str(j+1)

  return "unidentified"

print(search_matrix_1(N, K))
````

**Algoritma 2.1**

Best case = worst case
* Efisiensi waktu: O($n^{2}$)
* Efisiensi ruang: O($n^{2}$)

````python
# ALGORITMA 2.2

N = int(input("N = "))
K = int(input("K = "))

def search_matrix_2(N, K):
  counter = 1
  matrix = []

  # LAKBAN
  if K > N * N:
    return "unidentified"

  # building the matrix
  for i in range(N):
    # building the i-th row of the matrix
    row = []
    for j in range(N):
      # building each element of the i-th row
      row.append(counter)
      counter += 1
    # append the i-th row of the matrix
    matrix.append(row)

  # looking for the position of K in the matrix
  i, j = 0, 0

  # traversing row by row first
  while matrix[i][j] < K and i < N-1:
    i += 1
  # the correct row is found

  # traversing column by column on the correct row
  while matrix[i][j] < K:
    j += 1

  return "R = " + str(i+1) + " C = " + str(j+1)

print(search_matrix_2(N, K))
````

**Algoritma 2.2**

Best Case:
* Efisiensi waktu: $\Omega$(1)
* Efisiensi ruang: $\Omega$(1)

Worst Case:
* Efisiensi waktu: O($n^{2}$)
* Efisiensi ruang: O($n^{2}$)

````python
# ALGORITMA 2.3

N = int(input("N = "))
K = int(input("K = "))

def search_matrix_3(N, K):
  # looking for the position of K in the matrix
  i, j = 0, 0
  val = 1 # val in the value of matrix[i][j]

  # traversing row by row first
  while val < K and i < N-1:
    i += 1
    val += N
  # the correct row is found

  # traversing column by column on the correct row
  while val < K and j < N-1:
    j += 1
    val += 1

  if val != K:
    return "unidentified"

  return "R = " + str(i+1) + " C = " + str(j+1)

print(search_matrix_3(N, K))
````

**Algoritma 2.3**

Best Case:

* Efisiensi Ruang: $\Omega$(1)

Worst Case:
* Efisiensi Waktu: O(n)
* Efisiensi Ruang: O(1)

**Tantangan:**

**-> Algortima terbaik di antara 3 algoritma di atas adalah algoritma 2.3**

````python
# Efisiensi Waktu: O(log N)
# Efisiensi Ruang: O(1)

N = int(input("N = "))
K = int(input("K = "))

def search_matrix_Ologn(N,K):
    if K > N*N or N<1 or K<1:   # Tambal
        return "unidentified"
    else:
        column = K % N              
        if column == 0: 
            column = N

        # a akan menjadi patokan untuk mendapatkan batas kiri dan kanan
        if N % 2 == 0:
            a = N//2       
            row = N//2
        else:
            a = N//2 + 1
            row = N//2 + 1

        batas_kiri = (1 + N*a)-N
        batas_kanan = N*a

        while batas_kiri > K or batas_kanan < K:
            if a % 2 == 0:
                a = a//2
            else:
                a = a//2 + 1
            if batas_kiri > K:  # cari di baris atasnya
                if (batas_kiri - N*a) < 0:  # Kondisi jika batas < 0
                    batas_kiri = batas_kiri - N*(a//2)
                    batas_kanan = batas_kanan - N*(a//2)
                    row = row - a//2
                else:
                    batas_kiri = batas_kiri - N*a
                    batas_kanan = batas_kanan-N*a
                    row = row - a
            elif batas_kanan < K:   # cari di baris bawahnya
                if (batas_kanan + N*a) > N*N:   # Kondisi jika batas kanan > N*N 
                    batas_kiri = batas_kiri + N*(a//2)
                    batas_kanan = batas_kanan + N*(a//2)
                    row = row + a//2
                else:
                    batas_kiri = batas_kiri + N*a
                    batas_kanan = batas_kanan + N*a
                    row = row + a
    return f"R = {row} C = {column}"
print(search_matrix_Ologn(N,K))
````

````python
# Efisiensi Waktu: O(1)
# Efisiensi Ruang: O(1)

N = int(input("N = "))          # Matrix n x n
K = int(input("K = "))          # Bilangan yang ingin dicari

def search_matrix_O1(N,K):
    if K > N*N or N<1 or K<1:   # Tambal
        return "unidentified"
    else:
        kolom = K % N                           # Di modulo untuk mendapatkan kolom ke berapa
        if kolom == 0:                          # Jika hasil modulo = 0, seharusnya itu adalah kolom ke-N
            kolom = N                           # Ubah kolom menjadi kolom ke-N
        baris = (K+N-1)//N                      # Menentukan baris ke berapa 
        return f"R = {baris} C = {kolom}"

print(search_matrix_O1(N,K))
````

## Tugas 3 - Perancangan Algoritma

Diberikan input berupa sebuah matriks berukuran $N \times M$ dengan $N,M$ keduanya bilangan bulat positif. Matriks itu menggambarkan peta dari suatu lahan ranjau. Entri-entri matriks tersebut adalah suatu karakter ```B``` atau ```T```, secara berturut-turut merepresentasikan petak yang berisi bom dan petak yang berisi tanah biasa.

Pada waktu tertentu bom akan meledak dan akan menghancurkan seluruh petak lain dalam peta yang sebaris dan sekolom dengan posisi bom tersebut. Anda diminta untuk menghitung ada berapa banyak petak dalam peta tersebut yang akan tetap aman tanpa menerima ledakan dari bom-bom yang mungkin ada?


**Contoh.** Misalkan $M= 3, N=5$ dan matriks berisikan karakter-karakter sebagai berikut:
```
BTTBT
TTBTT
TTTTT
```
Maka setelah meledak, bom akan menghancurkan petak-petak yang ditandai dengan karakter ```*``` sebagai berikut:
```
*****
*****
*T**T
```
Artinya, setelah bom meledak, peta masih menyisakan 2 petak tanah yang masih aman dari jangkauan bom. Jadi, program akan mengoutputkan ```2```.


**Batasan Input.**
* Dijamin bahwa $M$ dan $N$ adalah bilangan bulat positif.
* Dijamin bahwa isian matriks hanya berisi 1 dari 2 karakter 'B' atau 'T' saja.

**Tugas.** Rancanglah suatu algoritma untuk menjawab permasalahan ini dengan batasan efisiensi waktu dan ruang $\mathbf{O}(n^3)$ dan $\mathbf{O}(n^2)$ secara berturut-turut.

**Tantangan.** Rancanglah suatu algoritma untuk menjawab permasalahan ini dengan batasan efisiensi waktu dan ruang keduanya dalam $\mathbf{O}(n^2)$.

````python
N = 5                   # total kolom
M = 3                   # total baris

matrix = []

for i in range(M):      # buat matrix N x M dengan isian T
    row_first_matrix = []
    for j in range(N):
        row_first_matrix.append("T")
    matrix.append(row_first_matrix)

# Misal ada 3 bom
matrix[0][0] = 'B'      # baris 1 kolom 1
matrix[0][3] = 'B'      # baris 1 kolom 4
matrix[1][2] = 'B'      # baris 2 kolom 3

def bomb(matrix):
    bomb_row = []
    bomb_column = []

    row_count = 0           
    bomb_count = 0
    not_destroyed_count = 0

    for row in matrix:
        for j in range(N):
            if row[j] == 'B':
                bomb_row.append(row_count)       # Menyimpan index baris bom
                bomb_column.append(j)            # Menyimpan index kolom bom
                bomb_count += 1                  # Menghitung berapa bom
        row_count += 1

    for i in range (bomb_count):
        row_bomb = bomb_row[i]                   # Ambil index baris bom ke-i
        column_bomb = bomb_column[i]             # Ambil index kolom bom ke-i

        # Peledakan kolom:
        for j in range(M):
            matrix[j][column_bomb] = '*'

        # Peledakan baris:
        for k in range(N):
            matrix[row_bomb][k] = '*'

    for i in matrix:                        # Menghitung jumlah petak yang tidak terkena bom
        for j in range(N):
            if i[j] == 'T':
                not_destroyed_count += 1
    return not_destroyed_count

print(f"Ada {bomb(matrix)} daerah yang tidak terkena bom")
````
