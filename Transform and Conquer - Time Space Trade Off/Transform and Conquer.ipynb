### Tugas 1

Buatlah suatu algoritma untuk mendeteksi apakah dua buah string *saling anagram* atau tidak.

Suatu string ```s1``` merupakan sebuah anagram dari string ```s2``` jika setiap karakter di ```s1``` dapat disusun ulang urutannya sehingga menjadi ```s2```.

Contoh: string ```study``` adalah anagram dari string ```dusty```, namun string ```tolong``` bukan merupakan anagram dari string ```lontong```.

Ketentuan kompleksitas waktu: $\mathcal{O}(n\log n)$,

Ketentuan kompleksitas ruang: $\mathcal{O}(n)$,

di mana $n$ merupakan ukuran string terpanjang dari antara dua string yang sedang diuji.


````python
##### MERGE SORT #####
def pecah_jadi_dua(arr):
    n = len(arr)

    # Bagi 2 : left & right
    mid = n // 2
    left = arr[0:mid]
    right = arr[mid:n]

    return left, right

def gabung(left,right):
    final_list = []     # Tempat menampung list yang telah digabung

    # Loop sampai list kiri dan list kanan habis
    while len(left) != 0 and len(right) != 0:   
        # Lihat data yang lebih kecil untuk di append ke final_list
        if left[0] <= right[0]:
            data = left.pop(0)
            final_list.append(data)
        else:
            data = right.pop(0)
            final_list.append(data) 
        
        # Kalau list kiri sudah habis dan list kanan masih ada, masukin sisanya
        if len(left) == 0 and len(right) != 0:
            for data in right:
                final_list.append(data)

        # Kalau list kanan sudah habis dan list kiri masih ada, masukin sisanya
        if len(right) == 0 and len(left) != 0:
            for data in left:
                final_list.append(data)

    return final_list

def urut(arr):
    # Langkah basis
    if len(arr) == 1:
        return arr

    # Pecah jadi 2 terus
    kiri,kanan = pecah_jadi_dua(arr)
    kiri = urut(kiri)
    kanan = urut(kanan)

    # Gabungin hasil pecahannya dalam bentuk yang terurut
    arr = gabung(kiri,kanan)

    return arr

def string_sorting(_string) :
    string_list = list(_string)
    sorted_string = urut(string_list)
    return ''.join(sorted_string)

def is_anagram(string_1, string_2) :
    s1_length = len(string_1)
    s2_length = len(string_2)

    # Kalau panjangnya ga sama, udah pasti tidak anagram
    if s1_length != s2_length :
        return f"String '{string_1}' and '{string_2}' are not mutually anagrammatic!"

    else : 
        storage_counter = {}
        
        # Sort kedua string
        s1_sorted = string_sorting(string_1)
        s2_sorted = string_sorting(string_2)

        # masukin setiap karakter dari s1_sorted ke dictionary
        for char in s1_sorted:
            if char in storage_counter :
                storage_counter[char] += 1
            else :
                storage_counter[char] = 1

        # Kurangin karakter yang ada di s_1 sorted yang sama dengan s_2 sorted
        for char in s2_sorted:
            if char in storage_counter.keys() :
                storage_counter[char] -= 1

    # Cek key values
    for k in storage_counter.values() :
        if k != 0 :     # Kalau key valuesnya 0 semua berarti anagram
            return f"String '{string_1}' and '{string_2}' are not mutually anagrammatic!"
    return f"String '{string_1}' and '{string_2}' are mutually anagrammatic!"


s1 = "tolong"
s2 = "lontong"
angram = is_anagram(s1, s2)
print(angram)
````

````python
# H = [None, 2, 7, 3, 4, 10, 8, 9, 6, 17, 13, 5, 12, 15]    # Test Bottom Up
# H = [None, 2, 3, 14, 4, 10, 8, 9, 6, 17, 13, 15, 12, 18]  # Test Top Down
H = [None, 8, 2, 4, 9, 1, 7, 3, 10, 5]                    # Test Heapify
# H = [None, 1, 2, 3, 5, 8, 7, 4, 10, 9]                    # Test Insert
# H = [None, 1, 2, 3, 5, 6, 7, 4, 10, 9, 8]                 # Test Delete

def BottomUpParentDominance(i):
    # Tambal
    if i > len(H)-1 or i <1:
        return "Input tidak valid"
    
    # Basis
    if i == 1:
        return H
    
    child = H[i]        # angka childnya
    par = H[i//2]       # angka parentnya
    if par > child:     # cek apakah parental dominance
        H[i//2] = child     # Kalau tidak, tuker!
        H[i] = par
        BottomUpParentDominance(i//2)
    return H 

def TopDownParentDominance(i):
    # Tambal
    if i > len(H)-1 or i < 1:
        return "Input tidak valid"
    
    # kalau udah ga punya anak
    if 2*i > len(H)-1:
        return H

    # Ambil index anak anaknya
    LC_index = 2*i
    RC_index = 2*i+1

    # Ambil value anak kirinya
    LC = H[LC_index]

    # Kalau ada anak kanan:
    if RC_index <= len(H)-1:
        RC = H[RC_index]

        # Kalau ga parental dominance:
        if H[i] > LC or H[i] > RC:
            MinimumChild = min(LC,RC)  # Cari minimum child

            # Langkah SWAP
            if MinimumChild == LC:
                H[LC_index] = H[i]
                H[i] = LC
                MinimumChild_index = LC_index
            else:
                H[RC_index] = H[i]
                H[i] = RC
                MinimumChild_index = RC_index
            TopDownParentDominance(MinimumChild_index)
    # Kalau ga punya anak kanan
    else:
        # Bandingin sama anak kiri aja, kalau ga parental dominance, swap
        if H[i] > LC:
            H[LC_index] = H[i]
            H[i] = LC
            MinimumChild_index = LC_index
            TopDownParentDominance(MinimumChild_index)
    return H

def Heapify(H):
    N = len(H) - 1      # Banyak angka yang ada di list
    i = N//2            # Jumlah parent yang ada dalam list

    # Lakukan TopDown dari i sampai 1
    while i != 0:
        TopDownParentDominance(i)
        i -= 1
    return H

def Insert(v):
    # Masukkan nilai v ke list paling belakang
    H.append(v)
    new_val_index = len(H)-1
    # Lakukan bottom up dari index angka yang baru dimasukkan
    BottomUpParentDominance(new_val_index)
    return H

def Delete(i):
    # Tambal
    if i > len(H)-1 or i < 1:
        return "Input tidak valid"
    
    # Swap nilai dari indeks yang ingin di-delete dengan nilai dari indeks terakhir
    i_val = H[i]
    last_val = H[len(H)-1]

    H[i] , H[len(H)-1] = last_val , i_val

    # Hapus daun terakhir
    H.pop(-1)
    # Lakukan Top Down dari indeks yang baru saja ditukar dengan indeks terakhir
    TopDownParentDominance(i)
    return H

# print(TopDownParentDominance(3))
# print(BottomUpParentDominance(4))
print(Heapify(H))
# print(Insert(6))
# print(Delete(2))
````

**Tugas 3.3.** Temukan sebuah instance yang merupakan worst case dari algoritma Horspoll. Buktikan kompleksitas kasus terburuk tersebut.

````python
def create_shift_table(pattern, alphabet):
    m = len(pattern)
    shift_table = {}  # Membuat shift_table

    # Membentuk isi data dari shift_table dengan pengeseran awal bernilai m untuk semua alphabet
    for char in alphabet:
        shift_table[char] = m

    # Mengubah pengeseran tiap alphabet yang ada dalam pola tersebut
    for i in range(m - 1):
        char = pattern[i]
        shift_table[char] = m - 1 - i

    return shift_table

def horspool_search(text, pattern, shift_table):
    n = len(text)
    m = len(pattern)

    count = 0   # Untuk menghitung berapa kali iterasi

    i = 0   # Untuk mengetahui indeks string selanjutnya

    matches = [] # Menyimpan indeks pattern yang sesuai dengan substring

    # Looping metode horspool
    while i <= n - m:
        j = m - 1
        while j >= 0 and pattern[j] == text[i + j]:
            j -= 1

        # Pemeriksaan awal
        if count == 0:
            print("\nPemeriksaan awal:")
            print(text)
            print(" " * i + pattern)
            if j == -1:
                print(f"String ditemukan pada indeks ke {i}.")
                matches.append(i)
            count += 1
            shift = shift_table.get(text[i + m - 1])

            i += shift
            print()
            continue        
        
        # Kalau j = -1, berarti patternnya match
        if j == -1:
            print(f"shift #{count}:")
            print(text)
            print(" " * i + pattern)
            print(f"String ditemukan pada indeks ke {i}.\n")
            matches.append(i)
        else:   # Kalau engga, berarti patternnya ga sesuai
            print(f"shift #{count}:")
            print(text)
            print(" " * i + pattern)
            print()

        # shifting ke substring berikutnya
        shift = shift_table.get(text[i + m - 1])

        i += shift
        
        count += 1
    
    if len(matches) == 0:
        return -1
    else:
        string_data = ''
        for i in range(len(matches)):
            if i == len(matches) - 1:
                string_data += str(matches[i])
                break
            string_data += str(matches[i]) + ', '
        return f"String ditemukan pada indeks ke {string_data}."

def main():
    text = "SHDOKOSPJANHHAUVIISDJEANCHAUVINPOSLBSJEANCHAUVINSKFL"
    pattern = "JEANCHAUVIN"
    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    shift_table = create_shift_table(pattern, alphabet)

    print("Shift Table:")
    for char, shift in shift_table.items():
        print(f"| {char} | {shift:2} |", end="  ")

    print()
    searching = horspool_search(text, pattern, shift_table)
    print(searching)

if __name__ == "__main__":
    main()
````

Worst case terjadi bila seluruh karakter dari pattern dan text merupakan karakter yang sama seluruhnya. Hal ini mengakibatkan perpindahan atau shift menjadi satu persatu karena seluruh karakter sama disertai pemeriksaan masing-masing karakter dalam text sesuai dengan pattern.

### Tugas 4

Dalam tugas ini, Anda akan mengimplementasikan Algoritma Rabin-Karp untuk menyelesaikan masalah String Matching dalam konteks bioinformatika. 

Sebuah genome sequence merupakan sebuah string yang setiap karakternya dapat diambil dari himpunan karakter $\{\mathtt{A}, \mathtt{G}, \mathtt{T}, \mathtt{C}\}$ yang masing-masing mewakili 4 jenis basa yang menyusun molekul DNA.

Diberikan sebuah genome sequence yang disimpan oleh variabel string ```genome``` sepanjang $n$ dan sebuah pola basa nukleotida yang disimpan oleh variabel ```pattern``` sepanjang $m$. Anda diminta untuk mencari **semua** lokasi kemunculan ```pattern``` pada ```genome``` dengan menggunakan Algoritma Rabin-Karp.

Untuk setiap posisi pengecekan substring dari ```genome```, Anda perlu mengoutputkan setiap hash-key value dari substring tersebut.

Contoh:
```
genome = AGTCGTTAGAT
pattern = CGTTA
```
Kurang lebih diharapkan algoritma dapat menghasilkan output sebagai demikian.
```

```

*Catatan: Output di atas hanya salah satu metode dari berbagai cara mendesain fungsi hashingnya. Anda tidak harus mengikuti fungsi hashing seperti di atas.*

Ketentuan kompleksitas waktu: $\mathcal{O}(mn)$, dengan average case $\Theta(n)$.

Ketentuan kompleksitas ruang: $\mathcal{O}(m+n)$.

````python
def convert_to_num(char) :
  # Mengubah himpunan karakter {A,G,C,T} ke dalam bentuk angka
  char_to_num_dictionary = {
      'A' : 0,
      'G' : 1,
      'C' : 2,
      'T' : 3
  }
  if char in char_to_num_dictionary:
    return char_to_num_dictionary[char]
  else:   # Validasi
    return None   

def hash(text, k):
  result = 0
  m = len(text)

  # Mengkonversi string ke angka (hash)
  for i in range(m):
    con_num = convert_to_num(text[i])
    if con_num == None:   # Validasi
      return None
    result += (con_num * (k**(m - i - 1)))
  return result

def rehash(substring_code, old_char, new_char, k, m) :
  # Rehash: buang yang depan, * 5, tambahin di karakter baru di belakang
  rehash_code = (substring_code - convert_to_num(old_char) * (k**(m - 1))) * k + convert_to_num(new_char)
  return rehash_code

def Rabin_Karp(genome, pattern, k) :
  n, m = len(genome), len(pattern)  # menghitung panjang dari genome dan pattern
  code_p = hash(pattern, k) # pattern code

  # Validasi jika salah satu karakter dari pattern tidak ada di text keseluruhan
  if code_p == None:
    return "Invalid Pattern!"
  
  substring = genome[0:m]   # Ambil substring awal sepanjang m
  code_s = hash(substring, k)   # hash substring awal (dari text)

  # Pengecekan semua substring dengan pattern
  for j in range(n-m+1):
    print(f"Iterasi {j+1}. substring: {substring}, value = {code_s}")

    # Kalau patternnya ketemu, output ketemu!
    if code_s == code_p :
      print("Pattern is found!")

    # Kalau masih ada substring yang bisa diperiksa, lanjut ke sebelahnya!
    if j < n-m :
      substring = genome[j+1:m+j+1]   # substring baru
      old_char, new_char = genome[j] , genome[j+m]  # Identifikasi karakter yang mau diganti dan ditambahkan
      code_s = rehash(code_s, old_char, new_char, k, m)   # melakukan rehash untuk mengecek kode genome disebelahnya
      print()
  return "END"

def main():
  genome = "ACGTACGATCGATCGATGCAAACCCGTTTCTCT"
  pattern = "TTCT"
  k = 4
  result = Rabin_Karp(genome, pattern, k)
  print(result)

if __name__ == '__main__':
  main()
````
